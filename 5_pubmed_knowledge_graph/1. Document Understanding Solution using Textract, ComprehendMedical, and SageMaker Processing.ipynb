### SageMaker Processing

SageMaker Processing is used for this batch process. 
The input is a document in the PDF format. The output is a csv file describing entities and entity relationships found in the document. 

First, Amazon Textract is used as OCR in which printed texts on a PDF file is converted to strings inline by line. Detected lines are combined into one string per page. 
Second, Amazon Comprehend Medical is used as NER tool. Comprehend Medical returns a csv file per document describing a list of entities and entity-entity relationships. 

This batch process is further paralellized using multiple processing instances and multi process. 

### Modules

```python
import boto3
import pandas as pd
from sagemaker.processing import Processor, ProcessingInput, ProcessingOutput
from sagemaker import get_execution_role

my_session = boto3.session.Session()
my_region = my_session.region_name
my_account = boto3.client('sts').get_caller_identity().get('Account')
```

### Docker file

```python
docker_str = \
f"""
FROM nvidia/cuda:10.2-runtime-ubuntu18.04

RUN apt-get update \
  && apt-get -y install --no-install-recommends git python3 python3-pip libsm6 libxext6 \
    libxrender1 libglib2.0-0 python3-setuptools \
    ffmpeg libgl1-mesa-glx \
  && rm -rf /var/lib/apt/lists/*

RUN pip3 install --upgrade pip

RUN pip3 install boto3 pandas

WORKDIR /opt/ml/code
COPY doc2csv.py .
#COPY input.manifest .

# Make sure python doesn't buffer stdout so we get logs ASAP.
ENV PYTHONUNBUFFERED=TRUE
ENV PYTHONPATH=/opt/ml/code
ENV MY_REGION={my_region}
ENV BUCKET_NAME={bucket_name}
ENV S3_PREFIX={prefix}

ENV NVIDIA_VISIBLE_DEVICES all
ENV NVIDIA_DRIVER_CAPABILITIES compute,video,utility

ENTRYPOINT ["python3", "/opt/ml/code/doc2csv.py"]

"""
```

### Build a docker image

```python
with open("Dockerfile.part1", "w") as f:
    f.write(docker_str)
    
!docker build -t {repo_name} -f Dockerfile.part1 .
```

### Push to ECR

Attach AmazonEC2ContainerRegistryFullAccess to NeptuneML-HCLS-KG-Workshop-NeptunMLC-ExecutionRole-TKZBSPA4BLWQ

```python
%%bash -s {my_account} {my_region} {repo_name}

# If the repository doesn't exist in ECR, create it.
$(aws ecr get-login --no-include-email --region $2)

aws ecr describe-repositories --region $2 --repository-names $3 > /dev/null 2>&1
if [ $? -ne 0 ]
then
    aws ecr create-repository --region $2 --repository-name $3 > /dev/null
fi

docker tag $3:latest $1.dkr.ecr.$2.amazonaws.com/$3:latest
docker push $1.dkr.ecr.$2.amazonaws.com/$3:latest
```

### Run a SageMaker Processing job

```python
def run(instance_type, instance_count):
    processor = Processor(image_uri=f'{my_account}.dkr.ecr.{my_region}.amazonaws.com/{repo_name}:latest',
                                       role=get_execution_role(),
                                       instance_count=instance_count,
                                       base_job_name=f"{repo_name}-job",
                                       instance_type=instance_type)

        
    destination = '/opt/ml/processing/input/'
    pi = ProcessingInput(
                    source=f's3://{bucket_name}/{prefix}/', 
                    s3_data_type='S3Prefix',
                    s3_input_mode='File',
                    s3_data_distribution_type='ShardedByS3Key',
                    destination=destination)
        
    
    
    processor.run(inputs=[pi],
                  outputs=[ProcessingOutput(source='/opt/ml/processing/output', 
                        destination=f's3://{bucket_name}/stdized-data-new/comprehend_results/csv/',
                        output_name='output',
                        s3_upload_mode='Continuous')],
                  wait=False,
                  logs=False)
    
    return processor

run('ml.m5.4xlarge', 1)
```
